```python
#Zad 2 - BM
#Wypisuje łańcuch i położenie wzorca względem niego
def wypiszStan(położenieWzorca,lancuchZnakow,wzorzec):
    print(lancuchZnakow)
    for x in range(położenieWzorca):
        print(" ",end='')
    print(wzorzec)

lancuchZnakow = "ABABACBABAB"
wzorzec = "ABA"
znakiWzorca = list(tuple(wzorzec))
#Słownik z ostatnimi położeniami znaków
ostatniePolozenieZnakowWzorca = {}

#tablica ostatnich wystąpień
for x in znakiWzorca:
    #rfind znajduje ostatnie wystapienie
    ostatniePolozenieZnakowWzorca[x] = lancuchZnakow.rfind(x)
print(ostatniePolozenieZnakowWzorca)
#długość wzorca
dlugoscWzorca = len(wzorzec)
dlugoscLancucha = len(lancuchZnakow)
print("Długość wzorca: "+str(dlugoscWzorca))
print("Długość Łancucha znakowego: " + str(dlugoscLancucha))

położenieWzorca = 0
print(położenieWzorca)
iterator = położenieWzorca-1
while(położenieWzorca+iterator < dlugoscLancucha):
    iterator = dlugoscWzorca-1
    while( położenieWzorca+iterator < dlugoscLancucha and wzorzec[iterator]==lancuchZnakow[położenieWzorca+iterator] and iterator>-1  ):
        if(iterator==0):
            #znaleziono dopasowanie
            #print("L30:True")
            print("Dopasowanie na pozycji:" + str(położenieWzorca))
            wypiszStan(położenieWzorca,lancuchZnakow,wzorzec)
            
            #przesun na nowa pozycje o dlugosc wzorca
            położenieWzorca = położenieWzorca+dlugoscWzorca
            iterator=dlugoscWzorca
            print("Nowe położenie po poprawnym odkryciu: " + str(położenieWzorca))   
            wypiszStan(położenieWzorca,lancuchZnakow,wzorzec)
            
        
        if(not (lancuchZnakow[położenieWzorca] in ostatniePolozenieZnakowWzorca)):
        #jeśli sie nie zawiera to przesun o cala dlugosc wzorca
            położenieWzorca = położenieWzorca + dlugoscWzorca
            print("nie zawiera sie przesun do pozycji: "+str(położenieWzorca))
        iterator -= 1
    
    if(położenieWzorca+iterator < dlugoscLancucha and not (lancuchZnakow[położenieWzorca+iterator] in ostatniePolozenieZnakowWzorca)):
        #jeśli sie nie zawiera sie i jesli nie przekroczylismy dlugosci tekstu to przesun o cala dlugosc wzorca 
        położenieWzorca = położenieWzorca + dlugoscWzorca
        print("nie zawiera sie przesun do pozycji: "+str(położenieWzorca))
    elif(położenieWzorca+iterator+1 < dlugoscLancucha):
        print("Krok o jeden")
        położenieWzorca +=1
    wypiszStan(położenieWzorca,lancuchZnakow,wzorzec)
    
#Ostatnie położenie poza pętlą jest wypisywane ze wzgledu na strukturę kodu, ale nie jest już sprawdzane
print("Ostatnie położenie poza pętlą jest wypisywane ze wzgledu na strukturę kodu, ale nie jest już sprawdzane")
```

```python
#3 Liczby 10. na dowolny inny
#System rozróżnia małe i wielkie litery czy pozwala na system do 51 liczb

tab = "0123456789"+"abcdefghijkmnoprstuwz"+"ABCDEFGHIJKMNOPRSTUWZ"
print(len(tab))

#system = 2
#liczba = 200
#tabSys = tab[:system]
#print(tabSys)


def konwertuj(liczba,system,tabSys):
    wynik = ""
    wynikKolejnosc=""
    while(liczba!=0):
        reszta=liczba%system
        wynik += tabSys[reszta]
        #print(wynik)
        liczba=liczba//system
        #print(liczba)
    for x in reversed(wynik):
        wynikKolejnosc += x
    return wynikKolejnosc

print("Koncowe Wyniki:")
print("System 2: "+konwertuj(136,2,tab[:2]))
print("System 16:"+konwertuj(136,16,tab[:16]))
print("System 8:"+konwertuj(136,8,tab[:8]))
print("System 32:"+konwertuj(136,32,tab[:32]))
print("System 52:"+konwertuj(136,52,tab[:53]))
print("System 52:"+konwertuj(51,52,tab))
```
